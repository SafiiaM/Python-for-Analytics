```python
def Cpu_company (x):
    return x.split()[0]
```

```python
df['Cpu_Company']= df['Cpu'].apply(Cpu_company)
df
```

```python
def Mem(x):
    res = x.split()[0][:-2]
    if x.split()[0][-2:] == 'TB':
         res = float(res)*1024
    return int(res)
     
```

```python
df['Memory_Amount']=df['Memory'].apply(Mem)
df
```

```python
def Mem(x):
    x = x.replace(' ','')
    res = x.split('B')[-1]
    return res 
```

```python
df['Memory_type']=df['Memory'].apply(Mem)
df
```

```python
df.drop(columns= ['Memory', 'ScreenResolution'], inplace=True)
df
```

```python
from pandas.core.reshape.pivot import pivot_table
```

```python
df.pivot_table(
    columns='Cpu',
    index='Ram',
    values='Price_euros',
    aggfunc='mean'
)
```

```python
df.pivot_table(
    index='Cpu',
    columns='Ram',
    aggfunc='count',
    values='Price_euros'
) # 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ
```

```python
df.pivot_table(
    index='OpSys',
    columns='Memory_Amount',
    aggfunc='mean',
    values='Price_euros'
)
# 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти
```

```python
pd.crosstab(
    index=df["OpSys"],
    columns=df["Memory_Amount"],
    aggfunc="mean",
    values=df["Price_euros"]
)
```
