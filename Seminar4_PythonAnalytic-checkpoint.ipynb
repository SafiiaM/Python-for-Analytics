```python
# Готовим данные для графика
data = df['Memory_Type'].value_counts()
data.index = ['SSD', 'HDD', 'FlashStorage', 'Hybrid']
data.head()
```

```python
memory = df['Memory_Type'].value_counts()
memory.head()

plt.figure(figsize=(6, 4))

plt.pie(memory.values, autopct='%1.1f%%', labels=memory.index)

plt.title('Распределение типов носителя')
plt.legend(memory.index);
```

```python
plt.figure(figsize=(12, 8))

plt.pie(memory.values, autopct='%1.1f%%', labels=memory.index, wedgeprops=dict(width=0.5))

plt.title('Доля типов носителей')
plt.legend(memory.index);
```
