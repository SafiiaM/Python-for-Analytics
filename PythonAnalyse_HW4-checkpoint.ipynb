```python
data = df['waterfront'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.bar(data['index'], data['waterfront'])
plt.title('Распределение домов от наличия вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Кол-во домов')
plt.xticks(rotation=30);
```

2.1. Основное количество домов имеет максимум чуть больше четверти ограниченного вида на набережную. С хорошим видом на набережную очень малое количество домов - до 1000 домов 

```python
data = df['floors'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.bar(data['index'], data['floors'])
plt.title('Распределение этажей домов')
plt.xlabel('Этажность')
plt.ylabel('Кол-во домов')
plt.xticks(rotation=30);
```

2.2. Более 10000 домов находятся в диапазоне от 0,6 до 1,4 этажей, также от 2000 до 8000 тысяч домов находятся в диапазоне от 1,6 до 2,4 этажей 

```python
data = df['condition'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.bar(data['index'], data['condition'])
plt.title('Распределение этажей домов')
plt.xlabel('Состояние домов')
plt.ylabel('Кол-во домов')
plt.xticks(rotation=30);
```

```python
# Готовим данные для графика
data = df['condition'].value_counts()
data.index = ['1=bad', '2=acceptable', '3=good', '4=very good', '5=super']
data.head()
```

```python
condition = df['condition'].value_counts()
condition.head()

plt.figure(figsize=(6, 4))

plt.pie(condition.values, autopct='%1.1f%%', labels=data.index)

plt.title('Распределение домов в зависимости от состояния')
plt.legend(data.index);
```

64,9% - домов в плохом состоянии, лишь 0,8% домов в состоянии "супер"
